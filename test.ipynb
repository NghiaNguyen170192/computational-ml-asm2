```python
from newsapi import NewsApiClient
import pandas as pd

# Initialize API
api = NewsApiClient(api_key='48e165ad92a546ff97484100870dd721')

btc_sources = [
    'ars-technica',
    'bloomberg',
    'business-insider',
    'crypto-coins-news',
    'engadget',
    'fortune',
    'hacker-news',
    'recode',
    'techcrunch',
    'techradar'
]
```

```python
from newspaper import Article

source_str = ','.join(btc_sources)

response = api.get_everything(
    q='bitcoin OR btc',
    sources=source_str,
    language='en',
    sort_by='relevancy',
    page_size=100,  # max per request,
)

articles_df = pd.DataFrame(response['articles'])

def extract_full_text(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        return article.text
    except Exception as e:
        return None

# Step 6: Apply extraction to each article
articles_df['full_content'] = articles_df['url'].apply(extract_full_text)

# Optional: Wait between requests to avoid being blocked
# time.sleep(1) inside extract_full_text if needed

# Step 7: Save to local file
# Option A: Save as CSV
# articles_df.to_csv('bitcoin_articles_full.csv', index=False, encoding='utf-8')

# Option B: Save as JSON
articles_df.to_json('bitcoin_articles_full.json', orient='records', lines=True)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('raw-data.csv')

# Convert all text to lowercase for consistent matching
df['title'] = df['title'].astype(str).str.lower()
df['description'] = df['description'].astype(str).str.lower()
df['content'] = df['content'].astype(str).str.lower()

# Define keywords
keywords = ['bitcoin', 'btc']

# Filter rows where any column contains the keywords
filtered_df = df[
    df['title'].str.contains('|'.join(keywords)) |
    df['description'].str.contains('|'.join(keywords)) |
    df['content'].str.contains('|'.join(keywords))
]

# Preview the filtered results
filtered_df[['title', 'description', 'content']].head()
```
